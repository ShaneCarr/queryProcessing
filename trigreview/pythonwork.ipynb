```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Circle

angle_degrees = 30
# radian is radius wrapped around circle
# 360 degress = 2*pi radians

# conversion factor
# the key is that both radians and derees are a measure of movement around a circle.
# degrees to radians: 360/ 2*pi * radians = 180/ (pi  * radians) --degrees per radian or vice versal
# this simple ratio can be wroked either way to convert from one to the other.
angle_radians = np.radians(angle_degrees)

opposite = np.sin(angle_radians)
adjacent = np.cos(angle_radians)
hypotenuse = 1
plt.figure(figsize=(10, 10), dpi=100)
####plt.figure(figsize=(5, 5))
# plt.plot([x1, x2], [y1, y2], 'k-')
plt.plot([0, adjacent], [0, 0], 'r-', label='Adjacent (cos)')
plt.plot([adjacent, adjacent], [0, opposite], 'b-', label='')
plt.plot([0, adjacent], [0, opposite], 'g--', label='Hypotenuse')

# draw the axis
plt.plot([0, 0], [0, 1], 'k--', alpha=0.3)  # Extend y-axis visually
plt.plot([1, 0], [0, 0], 'k--', alpha=0.3)
plt.plot([0, 0], [0, -1], 'k--', alpha=0.3)
plt.plot([-0, 0], [0, 0], 'k--', alpha=0.3)

# Add labels and legend
plt.text(adjacent / 2, -0.05, 'cos()', horizontalalignment='center')
plt.text(adjacent + 0.05, opposite / 2, 'sin()', verticalalignment='center')
plt.text(adjacent / 2 - 0.1, opposite / 2, '', horizontalalignment='center')
plt.legend(loc='upper right')

circle = Circle((0, 0), 1, fill=False, color='blue', linestyle='-', alpha=0.5)
plt.gca().add_patch(circle)
plt.xlim(-1.1, 1.1)
plt.ylim(-1.1, 1.1)
plt.gca().set_aspect('equal', adjustable='box')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Unit Circle with Trigonometric Functions')
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a range of angles from 0 to 2*pi radians
theta = np.linspace(0, 2 * np.pi, 100)

# Calculate x and y coordinates of points on the circle
x = np.cos(theta)
y = np.sin(theta)

# Plot the circle
plt.figure(figsize=(5, 5))
plt.plot(x, y)
plt.title('Unit Circle')
plt.xlabel('Cosine (x)')
plt.ylabel('Sine (y)')
plt.grid(True)
plt.axis('equal')  # Ensure the aspect ratio is equal to make the circle look round
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate an array of angles from 0 to 2*pi (360 degrees)
angles = np.linspace(0, 2 * np.pi, 400)

# Compute sine and cosine of these angles
sine_wave = np.sin(angles)
cosine_wave = np.cos(angles)

# Plotting sine and cosine waves
plt.figure(figsize=(10, 5))
plt.plot(angles, sine_wave, label='Sine Wave')
plt.plot(angles, cosine_wave, label='Cosine Wave', linestyle='--')

# Adding labels and legend
plt.xlabel('Angle (radians)')
plt.ylabel('Function value')
plt.title('Sine and Cosine Waves')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Define the sides of the triangle
a, b = 3, 4
c = (a**2 + b**2)**0.5  # Hypotenuse, c

# Create the plot
fig, ax = plt.subplots()
ax.set_aspect('equal')
plt.axis('off')

# Calculate positions for the large square
large_square_side = a + b

# Outer large square (a+b) x (a+b)
outer_square = patches.Rectangle((0, 0), large_square_side, large_square_side, linewidth=2, edgecolor='black', fill=False)
ax.add_patch(outer_square)

# Inner square a^2
square_a = patches.Rectangle((b, 0), a, a, linewidth=2, edgecolor='blue', fill=True, alpha=0.5)
ax.add_patch(square_a)

# Inner square b^2
square_b = patches.Rectangle((0, a), b, b, linewidth=2, edgecolor='green', fill=True, alpha=0.5)
ax.add_patch(square_b)

# Central square c^2, formed by the hypotenuses of the triangles
central_square = patches.Rectangle((b, a), c, c, linewidth=2, edgecolor='red', linestyle='--', fill=False)
ax.add_patch(central_square)

# Add the four right-angled triangles
triangles = [
    patches.Polygon([(b, a), (b, a+c), (b+c, a+c)], closed=True, fill=True, color='grey', alpha=0.5),  # Top right
    patches.Polygon([(b, a), (b+c, a), (b+c, a+c)], closed=True, fill=True, color='grey', alpha=0.5),  # Bottom right
    patches.Polygon([(0, a), (b, a), (b, 0)], closed=True, fill=True, color='grey', alpha=0.5),  # Bottom left
    patches.Polygon([(0, a), (0, a+b), (b, a+b)], closed=True, fill=True, color='grey', alpha=0.5),  # Top left
]

for t in triangles:
    ax.add_patch(t)

# Annotations for side lengths
plt.text(b/2, -0.2, 'b', ha='center', fontsize=12)
plt.text(b + a/2, -0.2, 'a', ha='center', fontsize=12)
plt.text(b + a + 0.2, a/2, 'a', va='center', fontsize=12, rotation='vertical')
plt.text(b - 0.2, a/2, 'b', ha='center', fontsize=12, rotation='vertical')
plt.text(b + a + c/2, a + c/2, 'c', color='red')

# Adjusting the plot limits
plt.xlim(-1, large_square_side + 1)
plt.ylim(-1, large_square_side + 1)

# Add a title and a key to the plot
plt.title('Pythagorean Theorem Visual Proof', fontsize=16, y=1.05)
plt.text(large_square_side + 1, large_square_side/2,
         "Key:\n\n"
         "Large outer square side: a + b\n\n"
         "Blue square area: a^2\n\n"
         "Green square area: b^2\n\n"
         "Red dashed square area: c^2\n\n"
         "Grey triangles: 4 right-angled triangles\n\n"
         "Proof:\n"
         "Area of large square = (a+b)^2\n"
         "Area of large square = 2ab + c^2\n"
         "Therefore, a^2 + b^2 = c^2",
         fontsize=12)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Define the triangle sides
a, b = 3, 4
c = (a**2 + b**2)**0.5  # Hypotenuse

# Create the figure and axis
fig, ax = plt.subplots()
ax.set_aspect('equal')
plt.axis('off')

# Outer large square (a+b) x (a+b)
outer_square = patches.Rectangle((0, 0), a+b, a+b, linewidth=2, edgecolor='black', fill=False)
ax.add_patch(outer_square)

# Inner squares for a^2 and b^2, correctly sized and positioned
square_a = patches.Rectangle((0, 0), a, a, linewidth=2, edgecolor='blue', fill=True, alpha=0.5)
square_b = patches.Rectangle((a, a), b, b, linewidth=2, edgecolor='green', fill=True, alpha=0.5)
ax.add_patch(square_a)
ax.add_patch(square_b)

# Central square for c^2, formed by the hypotenuses of the triangles
central_square = patches.Rectangle((a, 0), b, b, linewidth=2, edgecolor='red', linestyle='--', fill=False)
ax.add_patch(central_square)

# Draw the four right-angled triangles
# Triangle at the top left
triangle1 = patches.Polygon([(0, a), (a, a), (0, b+a)], fill=True, color='grey', alpha=0.5)
# Triangle at the top right
triangle2 = patches.Polygon([(a, a+b), (a+b, a+b), (a+b, a)], fill=True, color='grey', alpha=0.5)
# Triangle at the bottom right
triangle3 = patches.Polygon([(a+b, 0), (a+b, a), (a, 0)], fill=True, color='grey', alpha=0.5)
# Triangle at the bottom left
triangle4 = patches.Polygon([(a, 0), (0, 0), (0, a)], fill=True, color='grey', alpha=0.5)

ax.add_patch(triangle1)
ax.add_patch(triangle2)
ax.add_patch(triangle3)
ax.add_patch(triangle4)

# Label the squares and triangles
plt.text(a/2, a/2, '$a^2$', ha='center', va='center', fontsize=12)
plt.text(a+b/2, a+b/2, '$b^2$', ha='center', va='center', fontsize=12)
plt.text(a+b/2, b/2, '$c^2$', ha='center', va='center', fontsize=12, color='red')

# Set the limits of the plot
plt.xlim(-1, 2*(a+b))
plt.ylim(-1, 2*(a+b))

plt.show()
```
